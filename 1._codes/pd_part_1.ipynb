## 8. Fetch ONE specific column from the dataset
#### Notes:
    - There are two ways to fetch single column from the dataset
         - Passing column name as it is
         - Passing column name as a list

```python
# passing column name as it is
data_csv["id"]
```

```python
# passing column name as a list
data_csv[["id"]].head(6)
```

#### To understand the difference, let's check the type of the columns first

```python
# type of column while passing 1 column name as it is for retriving
type(data_csv["id"])
```

```python
# type of column while passing column name as a list for retriving
type(data_csv[["id"]])
```

#### ***Findings: 
- If we pass only a column name not as a list, then the type of that column is Series type. 
- And If we pass the column name as a list then the type of column will be DataFrame.

#### What if we want to fetch more than one column but not as a list, let's see what happens

```python
# fetching more than one column not as a list
# data["id", "address_1"] # it will give a Key error
```

```python
# fetching more than one column as a list
data_csv[["id", "address_1"]].head(8)
```

#### ***Findings: 
- We cen fetch a single column as a Series data type or as a DataFrame data type. 
- But, for fetching multiple column we MUST pass column name inside a list. 
- DataFrame is the collection of Series data type columns. That means a single column is of Series datatype
- Series is like a list. But, Series is from Pandas concept and list is from python core concept

## 12. How to rename column name using pandas?

```python
# rename column names
pd.read_excel("sample_excel/airline.xls", sheet_name = "Sheet2", names = ["a", "b", "c", "d", "e", "f"])
```

#### Notes:
- While renaming column, make sure you keep number of names equals to number of columns
- If #column names != #columns, then naming will align from right side of the table. Let's see what does it mean

```python
# rename column names. But, #column names != #columns
pd.read_excel("sample_excel/airline.xls", sheet_name = "Sheet2", names = ["a", "b", "c", "d"])
```

#### *** Findings:
- It seems like from above table index is missing!!
- But, interestingly, first two column is acting as index. And they are called multi layer indexes.
- Index is nothing but just a pointer to the particular row
