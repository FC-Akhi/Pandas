#### When we do merge operation:

- Whenever we bring datasets from different different sources, we need to check if data in one table is present in other table
- Or may be if we want to establish relationship between two data table we need above merge operations. For example two data table is given, left table is student table and right table is student access to course info table, if we want to find students who have access to course or not we can apply merge operations here to find the answers.
- **Important point:** Merge operation will not work when two table have different column names.

## 14. Difference between merge, join & concatenate
   
   - **Merge operation**
       - Merge operation is done based on **column values matching**
       - **Column names of two table must be same**
   #
   - **Join operation**
       - Join operation is done based on **row index matching**
       - **Column names of two table must be different**
   #
   - **Concatenate operation**
       - Concatenate operatin can join two table row wise
       - Or it can join two table column wise
