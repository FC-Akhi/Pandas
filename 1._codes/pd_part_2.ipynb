### Know the meaning of the above dataset
- Above dataset is the dataset of Titanic passengers. 
- **PassengerId:** Represents id of individual passenger
- **Survived:** THis column data represents whether a passenger survived or not
    - 1 = Survived 	&#x1F603;
    - 0 = Died &#x1F614;
- **Pclass:** It represents class of passenger ticket
    - 1 = class 1 ticket
    - 2 = class 2 ticket
    - 3 = class 3 ticket
- **Name:** This column represents name of a passenger
- **Sex:** This column represents gender of a passenger
- **Age:** This column represents age of a passenger
- **Ticket:** This column represents ticket name of a passenger
- **Fare:** This column represents fare a passenger paid
- **Cabin:** This column represents whether a passenger have cabin feature or not

- We have to do variety of statistical analysis on dataset and find different meaning from it.

## 4. Get brief statistical analysis of dataset

```python
# get statistical analysis of dataset
data_csv_git.describe()
```

#### Notes:
- Observe, **statistical analysis is given for only numerical format dataset**. Not for string/object/categorical format dataset

#### What each statistical component represents?
- **count:** Total number of entries. Here, there are 891 records.
    - Analysis per column:
        - Passengerid, Survived, Pclass, SibSp, Parch & Fare have complete 891 records
        - Age have (891 - 714 = 177) missing records or may be NaN
        - Depending on counts we decide whether we will keep the records or delete it
        - We may try to do fill operations like; average fill, backward or forward fill(will learn later). How to fill we will learn in statistics. So no worries...	&#x1F60A;
        - Whatever operations we will do on dataset, whether delete or fill it depends on type of business. Its totally contextual
- **mean:** Gives average of records to a specific columns. For example: avg age of a pessenger is 29.699, avg revenue or avg fare is 32.2
- **std:** std calculates, from mean/avg how much deviation do we have
- **min:** Determine minimum number from the records. For example; minimum age is 0.42(4 month infant), minimum of fare is 0 that means some person may received ticket as prize. And didn't have to pay for fare &#x1F603;
- **25%:** Read as 25th percentile. It reads record at 25th percentile(see visual representation below)  
- **50%:** Read as 50th percentile. It reads record at 50th percentile  
- **75%:** Read as 75th percentile. It reads record at 75th percentile  
- **max:** Determine maximum number from the records. For example; maximum age is 80, maximum of fare is 512.39!!

#### Notes: 
- To get whatever record is in 25, 50 or 75%, first we need to arrange records in ascending order. Then find what is present at 25th percentile or 50th percentile. What is the use of this information??
- We need this information so that we can understand at what range majority of the data exist.
- When we will have odd number of data, to get 25% we will take average of two numbers as 25%
- **Mean is definitely not equals to 50th percentile**

#### 5.1.Q: Above we wrote all categorical column names inside a list to get the description of these columns. What if we have 200 categorical columns. Are we able to write all column names manually &#x1F625; , NO. So, how could we solve this problem? Let's see...

## 6. Retrieve categorical column names without writing names manually

```python
# let's find data type of each columns again
data_csv_git.dtypes
```

```python
# find data type of output of dtypes
type(data_csv_git.dtypes)
```

#### Series!!!
#### Notes:
- **We know, Series is almost equivalent to a list. And, in case of series we will get index number or name**
- From above series we observe that PassengerId, Survived, Pclass, Name....Embarked are the index names. And actual data is int64, float64, object.

#### 6.1.Q: Why we need this above series, in which our column names from dataset became index? Let's see why...

```python
# we will try to get all columns which is of type object
data_csv_git.dtypes == "object"
```

#### Notes:
- Here, we used a comparison operator, So where ever it will find object it will give True otherwise False
- **We have to filter out data where it is True. Let's see how to do that.**

```python
# filter out only those records where this condition (data_csv_git.dtypes == "object") is TRUE
data_csv_git.dtypes[data_csv_git.dtypes == "object"]
```

#### WOW!! We filtered only object/string/categorical data. Bravo!! &#x1F44F; &#x1F483;

#### Notes: 
- Understand we could filter out all object type column name. BUT...
- What it's give us is a Series. Let's check type of (*data_csv_git.dtypes[data_csv_git.dtypes == "object"]*)

```python
# check type of data_csv_git.dtypes[data_csv_git.dtypes == "object"]
type(data_csv_git.dtypes[data_csv_git.dtypes == "object"])
```

#### Notes:
- In Above series, index names are Name, Sex, Ticket, Cabin & Embarked
- "object" is acting as actual data

#### But we want object type column names in a list. Let's see how to do that.

```python
# just print all index names from above Series. And we will get column names in a list
data_csv_git.dtypes[data_csv_git.dtypes == "object"].index # SEE HERE, WHEN WE GIVE .index IT IS GIVING COLUMN NAMES AS INDEX
```

```python
# pass above column names in data variable data_csv_git
data_csv_git[data_csv_git.dtypes[data_csv_git.dtypes == "object"].index]
```

#### PHEW!! That's lot of work we did..
