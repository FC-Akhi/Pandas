## 8. Where Age is < 30, keep record as it is. And when Age is => 30 replace it with logorithmic of respective Age value

```python
# first let us write the custom function
def cust(x):
    """This function will take Age as input
       Return: value depends on condition"""
    
    # when x = Age is smaller than 30
    if x < 30:
        return x
    
    # when x = Age is equal or greater than 30
    else:
        return math.log10(x)
```

```python
# create new column
data_csv_git_sub_1["Age_new"] = 0
```

```python
# show dataset
data_csv_git_sub_1
```

```python
# solution to this problem
data_csv_git_sub_1["Age_new"] = data_csv_git_sub_1["Age"].apply(cust)
```

```python
# show dataset
data_csv_git_sub_1
```

## 9. Where Sex == female, convert value to 1 and where Sex == male, convert it to 0

```python
# show dataset
data_csv_git_sub_1
```

```python
# first let us write the custom function
def func(x):
    """This function will take gender(female/male) as input
       Return: value depends on condition"""
    
    # when x = female
    if x == "female":
        return 1
    
    # when x = male
    else:
        return 0
```

```python
# solution to this problem
data_csv_git_sub_1["Sex_new"] = data_csv_git_sub_1["Sex"].apply(func)
```

```python
# show dataset
data_csv_git_sub_1
```

## 11. Seperate the entire dataset into three categories based on range of fare passengers paid
    
   - For example:
        - Category A is when fare is from 0 to 100
        - Category B is when fare is from 101 to 199
        - Category C is when fare is from 200 or +

```python
# first let us write the custom function
def func(x):
    """This function will take fare paid as input
       Return: value depends on condition"""
    
    # when fare is from 0 to 100
    if 0 <= x <= 100:
        return "A"
    
    # when fare is from 101 to 199
    elif 101 <= x <= 199:
        return "B"
    
    # when fare is 200 or +
    else:
        return "C"
```

```python
# solution
data_csv_git["Fare_group"] = data_csv_git["Fare"].apply(func)
```

```python
# show the dataset
data_csv_git
```

## 12. Filter out passengers record who paid between 101 and 199

```python
# filter according to the range of given fare
data_csv_git[data_csv_git["Fare_group"] == "B"] # we know from point 11, Category B is when fare is from 101 to 199
```
