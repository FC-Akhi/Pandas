#### Notes:
- While droping a column as we did above, it will give a key error. Please check by uncommenting
- What is the error??
    - ["Pclass&Age"] can't find in the axis. Well well, **What is AXIS**

- Our table dataset is 2 dimensional:
    - Row: Row is represent as Axis 0
    - Column: Column is represent as Axis 1
    
- From the error we see, drop method is searching our column name by default in some axis where it is absent. Check the by default value. It's axis = 0 which represents ROW. So, **let's specify axis...**

- Although we could see in outcome of drop column, that column is deleted. But, it didn't delete it permanently

Let's see why...

- Before that, let's see what other parameters present inside drop method


- There is a parameter called inplace.
- **What is inplace??**
    - Using inplace parameter we can control the deleteing operation. How??
    - If inplace = False, that means drop operation will be temporary
    - If inplace = True, that means drop operation will be permanent. And it will help to drop a column from original dataset



```python
# we can perform drop operation in two ways

# 1. reasign the drop operation to the dataset variable
# data_csv_git = data_csv_git.drop("Pclass&Age", axis = 1)

# 2. drop a column permanently using inplace = True
data_csv_git.drop("Pclass&Age", axis = 1, inplace = True)
```

## 16. Extract nth row using loc

```python
# fetch 5th row
data_csv_git.loc[4] # we should pass 5 not 4. See explanation below
```

```python
# show original dataset
data_csv_git.head(6)
```

#### Notes:
- Our requirement was to fetch 5th record. As we see from last point to fetch 5th record using iloc we need to pass (row number - 1) inside iloc. 
- In loc while passing (row number - 1 ) = (5 - 1) = 4, we are getting 4th index row from the table.
- There subtle difference between iloc and loc. To understand that let's understand types of indexing.

#### Types of indexing:
- There are two types:
    - 1. **System index**: What we not see, inside system is system index. Starts with 0th index 
    - 2. **Name index**: What we see while we read data table on screen. Starts with 1st index
    
#### Difference between iloc & loc with respect to row (Yes, there are difference between them with respect to row & column)
   - **iloc**: This will extract data according to system index. Say, extracting data from row 7th. 
    System index is 7 - 1 = 6th row 
   - **loc**: This will fetch data according to name index. Say, extracting data from row 7th.
    Name index is also 7 :)
    
    ** See difference with respect to column in point20 **

```python
# extract 5th row
data_csv_git.loc[5]
```

```python
# show original dataset
data_csv_git.head(6)
```

#### According to name index we extract 5th row :)

## 28. Explore parameter thresh in dropna() method for column axis or axis = 1

```python
# let's use dropna() again on subset data table and axis = 1
data_csv_git_sub.dropna(axis = 1)
```

#### Notes:
- What above column is doing is removing that column which have any NaN values.
- But, What if we have a condition like, drop a column when there is minimum two **non NaN** value. Otherwise not. Let's see how.
- To do that we need thresh parameter

```python
# show subset data table
data_csv_git_sub
```

```python
# drop column when thresh = 2
data_csv_git_sub.dropna(axis = 1, thresh = 2) 
```

#### Interestingly, in above table all column have more than two NaN values. Let's try thresh = 3

```python
# drop column when thresh = 3
data_csv_git_sub.dropna(axis = 1, thresh = 3) 
```

#### I think now you will have some explanation clear. When thres = 3, that means there should maximum 3 non NaN value in a column, otherwise column will be dropped. Let's try using thresh = 4

```python
# drop column when thresh = 4
data_csv_git_sub.dropna(axis = 1, thresh = 4) 
```

#### In Cabin column there was maximum 3 non NaN value. When we gave thresh = 4, it checked whether there is a column that has maximum 4 non NaN value or not, And it found that Cabin column have only 3 non NaN values. So dropna dropped Cabin column. I hope thresh concept is clear:)

#### Or in other words, atleast 4 non NaN values are required in a column, if not dropna will drop it.
